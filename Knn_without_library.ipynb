# **2C Pemodelan**

## 1. Inisialisasi K cluster

*   Secara acak pilih K titik awal dari data sebagai centroid.




```python
import numpy as np

def initialize_centroids(data, k):
    indices = np.random.choice(data.shape[0], k, replace=False)
    return data[indices]

# Pastikan df_scaled dalam bentuk array
data_array = df_scaled.to_numpy()
k = 3  # contoh: jumlah cluster

centroids = initialize_centroids(data_array, k)
print("Inisialisasi Centroid:")
print(centroids)
```

## 2. Assign setiap data ke cluster terdekat
* Gunakan jarak Euclidean untuk menentukan titik ke centroid terdekat.



```python
def assign_clusters(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
    return np.argmin(distances, axis=1)

labels = assign_clusters(data_array, centroids)
print("Label hasil assign:")
print(labels)
```

## 3. Update Centroid

```python
def update_centroids(data, labels, k):
    return np.array([data[labels == i].mean(axis=0) for i in range(k)])

new_centroids = update_centroids(data_array, labels, k)
print("Update centroid:")
print(new_centroids)
```

## 4. Looping hingga konvergen

```python
def kmeans(data, k, max_iters=100, tol=1e-4):
    centroids = initialize_centroids(data, k)
    for i in range(max_iters):
        labels = assign_clusters(data, centroids)
        new_centroids = update_centroids(data, labels, k)
        print(f"Iterasi {i+1}:\nCentroids:\n{new_centroids}\n")
        if np.allclose(centroids, new_centroids, atol=tol):
            print(f"Konvergen pada iterasi ke-{i+1}")
            break
        centroids = new_centroids
    return labels, centroids
```

## Jalankan clustering

```python
final_labels, final_centroids = kmeans(data_array, k=3)

# Simpan label ke dataframe
df_result = df_scaled.copy()
df_result['Cluster'] = final_labels

# Tampilkan hasil akhir
print(df_result.head())
```

```python
# Eksperimen 2: Uji Tanpa dan Dengan Beberapa Fitur
# Misalnya:
# Coba buang fitur seperti NumWebVisitsMonth atau Z_CostContact jika tidak informatif.
# Cek apakah skor membaik.

import pandas as pd
import matplotlib.pyplot as plt
# Eksperimen 2: Uji Tanpa dan Dengan Beberapa Fitur
# Definisikan ulang dataframe awal sebelum drop kolom, agar bisa melakukan eksperimen dengan set fitur berbeda
df_original = pd.read_csv('marketing_campaign.csv', sep='\t')

# Handling missing value pada df_original
df_original['Income'] = df_original['Income'].fillna(df_original['Income'].median())

# Feature Engineering pada df_original (ulang prosesnya)
df_original['Dt_Customer'] = pd.to_datetime(df_original['Dt_Customer'], format='%d-%m-%Y')
df_original['Age'] = 2024 - df_original['Year_Birth']
df_original['Total_Children'] = df_original['Kidhome'] + df_original['Teenhome']
df_original['Is_Married'] = df_original['Marital_Status'].apply(lambda x: 1 if x in ['Married', 'Together'] else 0)
df_original['Income_Per_Dependent'] = df_original['Income'] / (1 + df_original['Total_Children'])
latest_date = df_original['Dt_Customer'].max()
df_original['Days_Since_Enrollment'] = (latest_date - df_original['Dt_Customer']).dt.days
df_original['Has_Dependents'] = df_original['Total_Children'].apply(lambda x: 1 if x > 0 else 0)
df_original['Is_Parent'] = df_original['Total_Children'].apply(lambda x: 1 if x > 0 else 0)
df_original['Is_Single_Parent'] = df_original.apply(lambda row: 1 if row['Marital_Status'] in ['Single', 'Divorced', 'Widow'] and row['Total_Children'] > 0 else 0, axis=1)
df_original['Total_Spending'] = df_original['MntWines'] + df_original['MntFruits'] + df_original['MntMeatProducts'] + df_original['MntFishProducts'] + df_original['MntSweetProducts'] + df_original['MntGoldProds']
df_original['Accepted_Campaigns_Count'] = df_original['AcceptedCmp1'] + df_original['AcceptedCmp2'] + df_original['AcceptedCmp3'] + df_original['AcceptedCmp4'] + df_original['AcceptedCmp5']
df_original['Total_Channel_Purchases'] = df_original['NumDealsPurchases'] + df_original['NumWebPurchases'] + df_original['NumCatalogPurchases'] + df_original['NumStorePurchases']

# One-Hot Encode categorical features pada df_original
df_original = pd.get_dummies(df_original, columns=['Education', 'Marital_Status'], drop_first=True)

# Definisikan daftar fitur yang ingin dicoba (eksperimen)
# Eksperimen 1: Menggunakan semua fitur yang relevan setelah engineering dan encoding
features_experiment_1 = [col for col in df_original.columns if col not in ['ID', 'Year_Birth', 'Kidhome', 'Teenhome', 'Dt_Customer', 'Z_CostContact', 'Z_Revenue']]

# Eksperimen 2: Membuang fitur seperti NumWebVisitsMonth atau Z_CostContact (Z_CostContact sudah dibuang di atas, fokus ke NumWebVisitsMonth dan fitur potensial lain yang kurang informatif)
features_experiment_2 = [col for col in features_experiment_1 if col not in ['NumWebVisitsMonth', 'NumCatalogPurchases']] # Contoh membuang dua fitur

# Lakukan clustering dan evaluasi untuk setiap set fitur
feature_sets = {
    "Semua Fitur Relevan": features_experiment_1,
    "Tanpa NumWebVisitsMonth dan NumCatalogPurchases": features_experiment_2
}

for name, features in feature_sets.items():
    print(f"\n--- Eksperimen: {name} ---")

    # Pilih fitur untuk eksperimen ini
    df_experiment = df_original[features].copy()

    # Feature Scaling untuk set fitur ini
    scaler_exp = StandardScaler()
    scaled_features_exp = scaler_exp.fit_transform(df_experiment)
    df_scaled_exp = pd.DataFrame(scaled_features_exp, columns=df_experiment.columns)

    # Lakukan Clustering KMeans (gunakan implementasi sklearn untuk kemudahan dan konsistensi)
    # Kita akan coba beberapa nilai k seperti di langkah sebelumnya
    best_k = -1
    best_silhouette_score = -1

    print("Mencari k optimal menggunakan Silhouette Score:")
    for k in range(2, 9):
        kmeans_exp = KMeans(n_clusters=k, random_state=42, n_init=10) # Tambahkan n_init
        labels_exp = kmeans_exp.fit_predict(df_scaled_exp)
        score_exp = silhouette_score(df_scaled_exp, labels_exp)
        print(f"k = {k} → Silhouette Score: {score_exp:.4f}")

        if score_exp > best_silhouette_score:
            best_silhouette_score = score_exp
            best_k = k

    print(f"\nSet Fitur: {name}")
    print(f"Nilai k terbaik berdasarkan Silhouette Score: {best_k}")
    print(f"Silhouette Score terbaik: {best_silhouette_score:.4f}")

    # Interpretasi Silhouette Score untuk k terbaik
    if best_silhouette_score > 0.7:
        print("Kualitas Cluster: Sangat baik")
    elif 0.5 <= best_silhouette_score <= 0.7:
        print("Kualitas Cluster: Baik")
    elif 0.3 <= best_silhouette_score < 0.5:
        print("Kualitas Cluster: Cukup")
    else:
        print("Kualitas Cluster: Buruk")
```

```python
# Eksperimen 3: Uji dengan Teknik Dimensionality Reduction (Opsional)
# Coba pakai PCA untuk menurunkan dimensi (misalnya ke 2D atau 3D), lalu clustering.
# Cocok untuk data berdimensi tinggi.

import pandas as pd
import matplotlib.pyplot as plt
print("\n--- Eksperimen 3: Uji dengan Teknik Dimensionality Reduction (PCA) ---")

# Pilih fitur yang sudah diskalakan untuk PCA
data_for_pca = df_scaled.copy()

# Lakukan PCA untuk menurunkan dimensi
# Coba ke 2D dan 3D
n_components_list = [2, 3]

for n_components in n_components_list:
    print(f"\nMelakukan PCA ke {n_components} komponen")
    pca = PCA(n_components=n_components)
    pca_result = pca.fit_transform(data_for_pca)

    # Buat DataFrame dari hasil PCA
    pca_cols = [f'PCA_{i+1}' for i in range(n_components)]
    df_pca = pd.DataFrame(pca_result, columns=pca_cols)

    print(f"\nDataFrame setelah PCA ({n_components}D):")
    print(df_pca.head())

    # Lakukan Clustering KMeans pada data hasil PCA
    print("\nMencari k optimal pada data hasil PCA menggunakan Silhouette Score:")
    best_k_pca = -1
    best_silhouette_score_pca = -1

    for k in range(2, 9): # Coba nilai k yang sama
        kmeans_pca = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels_pca = kmeans_pca.fit_predict(df_pca)
        score_pca = silhouette_score(df_pca, labels_pca)
        print(f"k = {k} → Silhouette Score (PCA {n_components}D): {score_pca:.4f}")

        if score_pca > best_silhouette_score_pca:
            best_silhouette_score_pca = score_pca
            best_k_pca = k

    print(f"\nHasil untuk PCA ke {n_components} komponen:")
    print(f"Nilai k terbaik berdasarkan Silhouette Score: {best_k_pca}")
    print(f"Silhouette Score terbaik: {best_silhouette_score_pca:.4f}")

    # Interpretasi Silhouette Score untuk k terbaik pada data PCA
    if best_silhouette_score_pca > 0.7:
        print("Kualitas Cluster (PCA): Sangat baik")
    elif 0.5 <= best_silhouette_score_pca <= 0.7:
        print("Kualitas Cluster (PCA): Baik")
    elif 0.3 <= best_silhouette_score_pca < 0.5:
        print("Kualitas Cluster (PCA): Cukup")
    else:
        print("Kualitas Cluster (PCA): Buruk")

    # OPSIONAL: Visualisasi hasil clustering pada data PCA 2D
    if n_components == 2:
        # Lakukan clustering lagi dengan k terbaik untuk visualisasi
        kmeans_pca_best = KMeans(n_clusters=best_k_pca, random_state=42, n_init=10)
        labels_pca_best = kmeans_pca_best.fit_predict(df_pca)

        plt.figure(figsize=(8, 6))
        plt.scatter(df_pca['PCA_1'], df_pca['PCA_2'], c=labels_pca_best, cmap='viridis', s=20, alpha=0.8)
        plt.title(f"Clustering Results on PCA {n_components}D (k={best_k_pca})")
        plt.xlabel("PCA 1")
        plt.ylabel("PCA 2")
        plt.colorbar(label='Cluster')
        plt.grid(True)
        plt.show()
```
