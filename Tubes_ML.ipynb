## Feature engineering

```python
# Usia dari 'Dt_Customer' dan 'Year_Birth'
df['Age'] = 2024 - df['Year_Birth']

# Total Children
df['Total_Children'] = df['Kidhome'] + df['Teenhome']

# Status perkawinan - Biner (Married/Together vs lainnya)
df['Is_Married'] = df['Marital_Status'].apply(lambda x: 1 if x in ['Married', 'Together'] else 0)

# Pendapatan per tanggungan
df['Income_Per_Dependent'] = df['Income'] / (1 + df['Total_Children'])

# Waktu sejak pembelian terakhir (dengan asumsi 'Dt_Customer' adalah tanggal pendaftaran dan 'Recency' adalah hari sejak pembelian terakhir)
# Jika 'Recency' berarti sesuatu yang lain atau tidak ada info tanggal lain, ini akan berbeda.
# Dengan asumsi 'Recency' memang hari sejak pembelian terakhir relatif terhadap tanggal analisis
latest_date = df['Dt_Customer'].max()
df['Days_Since_Enrollment'] = (latest_date - df['Dt_Customer']).dt.days

# Fitur yang menunjukkan apakah pelanggan memiliki tanggungan
df['Has_Dependents'] = df['Total_Children'].apply(lambda x: 1 if x > 0 else 0)

# Fitur yang menunjukkan apakah pelanggan adalah orang tua
df['Is_Parent'] = df['Total_Children'].apply(lambda x: 1 if x > 0 else 0)

# Fitur yang menunjukkan apakah pelanggan adalah orang tua tunggal (jika berlaku dari Marital_Status dan Total_Children)
# Ini adalah pendekatan yang sederhana, dengan asumsi 'Single', 'Divorced', 'Widow' dengan anak-anak adalah orang tua tunggal.
df['Is_Single_Parent'] = df.apply(lambda row: 1 if row['Marital_Status'] in ['Single', 'Divorced', 'Widow'] and row['Total_Children'] > 0 else 0, axis=1)

# Kategori tingkat pendidikan (nanti bisa dikodekan one-hot)
# Buat pemetaan untuk pemesanan potensial jika diinginkan, atau simpan sebagai objek untuk one-hot.
education_mapping = {
    'Basic': 0,
    '2n Cycle': 1,
    'Graduation': 2,
    'Master': 3,
    'PhD': 4
}
# Disimpan sebagai objek untuk saat ini, pengodean one-hot adalah tipikal untuk kategori nominal
df['Education_Level'] = df['Education'].map(education_mapping)

# Total spending pada product
df['Total_Spending'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']


# Rata-rata pengeluaran per kategori produk (dapat dilakukan nanti jika diperlukan, mungkin akan menghasilkan terlalu banyak fitur)
# Untuk menyederhanakannya, mari kita hitung rata-rata pengeluaran per total produk yang dibeli jika kita memiliki info tersebut,
# tetapi kita hanya memiliki pengeluaran per kategori. Total pengeluaran adalah awal yang baik Feature related to accepted campaigns
df['Accepted_Campaigns_Count'] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5']

# Fitur terkait saluran yang digunakan
df['Total_Channel_Purchases'] = df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']

# Check the new columns
df[['Age', 'Total_Children', 'Is_Married', 'Income_Per_Dependent',
          'Days_Since_Enrollment', 'Has_Dependents', 'Is_Parent', 'Is_Single_Parent',
          'Total_Spending', 'Accepted_Campaigns_Count', 'Total_Channel_Purchases']].head()

df.info()
```

# **2C Pemodelan**

## 1. Inisialisasi K cluster

*   Secara acak pilih K titik awal dari data sebagai centroid.




```python
import numpy as np

def initialize_centroids(data, k):
    indices = np.random.choice(data.shape[0], k, replace=False)
    return data[indices]

# Pastikan df_pca dalam bentuk array
data_array = df_pca.copy()
k = 3  # contoh: jumlah cluster

centroids = initialize_centroids(data_array, k)
print("Inisialisasi Centroid:")
print(centroids)
```

## 2. Assign setiap data ke cluster terdekat
* Gunakan jarak Euclidean untuk menentukan titik ke centroid terdekat.



```python
def assign_clusters(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
    return np.argmin(distances, axis=1)

labels = assign_clusters(data_array, centroids)
print("Label hasil assign:")
print(labels)
```

## 3. Update Centroid

```python
def update_centroids(data, labels, k):
    return np.array([data[labels == i].mean(axis=0) for i in range(k)])

new_centroids = update_centroids(data_array, labels, k)
print("Update centroid:")
print(new_centroids)
```

## 4. Looping hingga konvergen

```python
def kmeans(data, k, max_iters=100, tol=1e-4):
    centroids = initialize_centroids(data, k)
    for i in range(max_iters):
        labels = assign_clusters(data, centroids)
        new_centroids = update_centroids(data, labels, k)
        print(f"Iterasi {i+1}:\nCentroids:\n{new_centroids}\n")
        if np.allclose(centroids, new_centroids, atol=tol):
            print(f"Konvergen pada iterasi ke-{i+1}")
            break
        centroids = new_centroids
    return labels, centroids
```

## Jalankan clustering

```python
final_labels, final_centroids = kmeans(data_array, k=3)

# Simpan label ke dataframe
df_result = df_scaled.copy()
# Ensure the index of final_labels matches the index of df_result
df_result['Cluster'] = final_labels
# Tampilkan hasil akhir
print(df_result.head())
```

## OPSIONAL (Visualisasi cluster 2D)

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
pca_result = pca.fit_transform(data_array)

plt.scatter(pca_result[:, 0], pca_result[:, 1], c=final_labels, cmap='Set1')
plt.title("Visualisasi Hasil Clustering")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

# **2D** **Evaluasi**

keterangan

| Silhouette Score | Kualitas Cluster |
| ---------------- | ---------------- |
| > 0.7            | Sangat baik      |
| 0.5 – 0.7        | Baik             |
| 0.3 – 0.5        | Cukup            |
| < 0.3            | Buruk            |



```python
from sklearn.metrics import silhouette_score

# Gunakan data hasil scaling dan label hasil clustering
score = silhouette_score(df_pca, final_labels)
print(f"Silhouette Score: {score:.4f}")

# Interpretasi Silhouette Score
if score > 0.7:
    print("Kualitas Cluster: Sangat baik")
elif 0.5 <= score <= 0.7:
    print("Kualitas Cluster: Baik")
elif 0.3 <= score < 0.5:
    print("Kualitas Cluster: Cukup")
else:
    print("Kualitas Cluster: Buruk")

```

# 2E EKSPERIMEN

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

for k in range(2, 9):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(df_pca)
    score = silhouette_score(df_pca, labels)
    print(f"k = {k} → Silhouette Score: {score:.4f}")

# Interpretasi Silhouette Score
if score > 0.7:
    print("Kualitas Cluster: Sangat baik")
elif 0.5 <= score <= 0.7:
    print("Kualitas Cluster: Baik")
elif 0.3 <= score < 0.5:
    print("Kualitas Cluster: Cukup")
else:
    print("Kualitas Cluster: Buruk")
```
